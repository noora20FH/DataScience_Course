#1. MOBIUS STRIP

```python
theta = np.linspace(0, 2 * np.pi, 30)
w = np.linspace(-0.25, 0.25, 8)
w, theta = np.meshgrid(w, theta)
```

```python
phi = 0.5 * theta
# radius in x-y plane
r = 1 + w * np.cos(phi)
x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))

# triangulate in the underlying parametrization
from matplotlib.tri import Triangulation
tri = Triangulation(np.ravel(w), np.ravel(theta))
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles,
                cmap='Greys', linewidths=0.2);
ax.set_xlim(-1, 1); ax.set_ylim(-1, 1); ax.set_zlim(-1, 1)
ax.axis('off');
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a parameterization for the Möbius strip
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)
x = (1 + 0.5 * v * np.cos(u / 2)) * np.cos(u)
y = (1 + 0.5 * v * np.cos(u / 2)) * np.sin(u)
z = 0.5 * v * np.sin(u / 2)

# Plot the Möbius strip
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b', alpha=0.8)

# Set plot limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-1, 1])

# Show the plot
plt.show()
```

#2. SUPERVIDES AND UNSUPERVISED

##**Supervised Learning:**



Supervised learning entails the training of a model on a labeled dataset, where each instance in the training data is associated with a corresponding target or output. The primary objective is for the model to acquire the ability to establish a mapping from inputs to outputs, enabling it to make predictions on new, unseen data. During the training process, the model is provided with input-output pairs, and its internal parameters are adjusted based on the input to minimize the disparity between its predictions and the actual outputs. An illustrative example of supervised learning is found in image classification, where the labeled dataset comprises images paired with their corresponding labels (e.g., cat or dog). The model then learns to categorize new images by recognizing patterns derived from the training data.


##**Unsupervised Learning:**




Unsupervised learning involves training a model on an unlabeled dataset, where the algorithm endeavors to identify patterns, relationships, or structures in the data without explicit guidance on what specific characteristics to seek. One prevalent task in unsupervised learning is clustering, wherein the algorithm groups similar data points together based on a defined similarity measure. Another task is dimensionality reduction, where the objective is to decrease the number of features while retaining crucial information. As an example, in clustering, the algorithm might group similar customer purchasing behavior without prior knowledge of what precisely defines a "group." In summary, supervised learning operates with labeled data, focusing on predicting outputs, whereas unsupervised learning operates with unlabeled data, aiming to unveil patterns or relationships inherent in the data.
