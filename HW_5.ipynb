#1. Please utilize the U.S. birth population dataset(births.csv) to plot the following Figure. Additionally, mark significant holidays such as:
 'New Year's Day,'
'Independence Day,'
'Labor Day,'
 'Halloween,'
'Thanksgiving,' and 'Christmas.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Read the dataset
df = pd.read_csv('/content/drive/MyDrive/BigData/births.csv')

# Convert 'year', 'month', and 'day' columns to datetime
df['date'] = pd.to_datetime(df[['year', 'month', 'day']], errors='coerce')

# Drop rows with missing date values
df = df.dropna(subset=['date'])

# Set the date as the index
df.set_index('date', inplace=True)

# Calculate daily births
daily_births = df.resample('D').sum()['births']

# Group by day of the year and calculate the average daily births
average_daily_births = daily_births.groupby(daily_births.index.dayofyear).mean()

# Plotting the spectral plot
plt.figure(figsize=(12, 6))
average_daily_births.plot(marker='', linestyle='-', color='blue')

# Adding labels and title
plt.xlabel('Day of the Year')
plt.ylabel('Average Daily Births')
plt.title('Spectral Plot of USA Births by Day of the Year (1969-1988)')

# Customize x-axis ticks to show month names
day_ticks = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
month_names = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
plt.xticks(day_ticks, month_names, rotation=45)

# Mark significant holidays
holidays = {
    'New Year\'s Day': '01-01',
    'Independence Day': '07-04',
    'Labor Day': '09-01',
    'Halloween': '10-31',
    'Thanksgiving': '11-23',  # Assuming Thanksgiving is always on the 1st day of November
    'Christmas': '12-25'
}

for holiday, date_str in holidays.items():
    holiday_date = pd.to_datetime(date_str, format='%m-%d')
    plt.axvline(holiday_date.dayofyear, color='red', linestyle='--', label=holiday)


# Adding legend
plt.legend()

# Show the plot
plt.show()
```

#2. Please draw the plot of a sine and a cosine curve, with the major and minor ticks in multiples of  π/2  and  π/4.  as following Figure.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the x-axis range
x = np.linspace(0, 3 * np.pi, 1000)

# Calculate the sine and cosine values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plotting the sine and cosine functions
plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='Sine', color='blue')
plt.plot(x, y_cos, label='Cosine', color='green')

# Adding labels and title

plt.title('Sine and Cosine Functions')

# Customize x-axis ticks
x_ticks = [0, np.pi / 2, np.pi, 3 * np.pi / 2, 2 * np.pi, 5 * np.pi / 2, 3 * np.pi]
x_ticklabels = ['$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$', r'$\frac{5\pi}{2}$', r'$3\pi$']
plt.xticks(x_ticks, x_ticklabels)

# Set y-axis range
plt.ylim(-3, 3)

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
